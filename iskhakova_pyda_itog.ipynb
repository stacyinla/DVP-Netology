**С помощью статистических методов проверить взаимосвязи между переменными.**

```python
from scipy import stats
```

```python
#Проверим взаимосвязь переменных и наличия диагноза M с помощью t-test
# H0: Диагноз M не зависит от значения переменной i
# H1: Диагноз M зависит от значения переменной i

for i in data.columns[2:]:
    print(i)
    data_one = data.loc[data["diagnosis"] == "M", i]
    data_two = data.loc[data["diagnosis"] == "B", i]
    print(stats.ttest_ind(data_one, data_two))
```

Вывод: Вероятность получения таких различий между записями с двумя диагнозами (M/B) для всех параметров, кроме symmetry_se, fractal_dimension_se, при действительном отсутсвии таких различий в ген. совокупности составляет < 0.05
То есть для всех параметров i кроме symmetry_se, fractal_dimension_se есть различия по значению между записями с двумя диагнозами (M/B), для этих параметров отвергаем гипотезу H0.
